In order to use experience and last new job as numerical data

```python
data.loc[data['experience'] == '<1', 'experience'] = 0
data.loc[data['experience'] == '>20', 'experience'] = 21
data['experience'] = pd.to_numeric(data['experience'])

data.loc[data['last_new_job'] == 'never', 'last_new_job'] = 0
data.loc[data['last_new_job'] == '>4', 'last_new_job'] = 5
data['last_new_job'] = pd.to_numeric(data['last_new_job'])
```

Spliting the input as numerical and categorical datas

```python
data_num = data[["city_development_index", "training_hours", "experience", "last_new_job"]].values
data_cat = data[["city","gender","relevent_experience","enrolled_university","education_level","major_discipline","company_type","company_size"]]
```

Standardizing numerical data

```python
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()
data_num = sc.fit_transform(data_num)
```

One Hot Encoding categorical data

```python
from sklearn.preprocessing import OneHotEncoder

ohe = OneHotEncoder()

data_cat = ohe.fit_transform(data_cat).toarray()
```

One Hot Encoding target column

```python
y = data.iloc[:,13:14].values
y = ohe.fit_transform(y).toarray()
```

Spliting data as train and test data

```python
from sklearn.model_selection import train_test_split

X = np.hstack((data_cat,data_num))
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.2)
dim_X = X_train.shape[1]
dim_y = y_train.shape[1]
```

Training model

```python
import keras
from keras.models import Sequential
from keras.layers import Dense

model = Sequential()
model.add(Dense(16, input_dim=dim_X,activation='relu'))
model.add(Dense(12, activation='relu'))
model.add(Dense(dim_y, activation='softmax'))
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])

history = model.fit(X_train, y_train, epochs=100, batch_size=64)
```

Checking model's performance

```python
y_pred = model.predict(X_test)
#Converting predictions to label
pred = list()
for i in range(len(y_pred)):
    pred.append(np.argmax(y_pred[i]))
#Converting one hot encoded test label to label
test = list()
for i in range(len(y_test)):
    test.append(np.argmax(y_test[i]))

from sklearn.metrics import accuracy_score
a = accuracy_score(pred,test)
print('Accuracy is:', a*100)
```
